# Pipeline: (from chapter 2) (fit transform, x_reduced, all same thing):

A sequence of data processing components. Passing each component to the next, in a automated way of workflows.

The pipeline takes a list of name/estimators, and executes each process in turn and passing it to the next. Then call the fit_transform onto the training set to process it.

#### e.g. - feature scaling
full_pipeline = Pipeline([<br>
    ('imputer', SimpleImputer(strategy="median")),<br>
    ('attribs_adder', CombinedAttributesAdder()),<br>
    ('std_scaler', StandardScaler()),<br>
])

housing_prepared = full_pipeline.fit_transform(housing)

#### e.g. polynomial features
poly_scaler = Pipeline([<br>
    ('poly_features', PolynomialFeatures(degree=90, include_bias=False)),<br>
    ('std_scaler', StandardScaler())<br>
])

X_train_poly_scaled = poly_scaler.fit_transform(X_train)

### Advantages:
- keeps the processes separate - easy debuggins
- a modular structure that is simpler to import


Can be used to streamline the entire machine learning algorithm - import, data cleaning, EDA/feature engineering, model training and prediction and make the algorithm into a modular structure.

In the book it's only been used for data-preprocessing applications like feature scaling.

# Dimensionality Reduction

Very important. When training data takes too long reducing the unnecessry details can considerably speed it up. It is also useful for data visualization.

However, it does make system perform worse and pipeline more complex/harder to maintain. Should be used only after the original data training is too slow. Sometimes it can filter out noise and unnecessary details, but usually it won't and just speeds up training.

Main approaches - projection and Manifold Learning
Techniques - PCA, Kernel PCA, LLE

## The Curse of Dimensionality

In high dimensions, randomly chosen points are likely to be extreme and very far apart. Higher dimensions mean that instances are likely to be sparse, hence making new instances likely to be far away from training instances and making predictions less reliable.

Greater the dimension of training set, greater the risk of overfitting.

### Projection

In real life, many features are almost constant and others are correlated. Training instances actually lie within a much lower-dimensional subspace of the high-dimensional space.

Projection projects the training instance perpendicuarly onto the subspace, thus reducing the dataset's dimensionality.

Not always the best as sometimes subspaces twist and turn, and projection squashes different layers together.


### Manifold Learning

2D Manifold - a 2D shape that can be bent/twisted in a higher-dimensional space. A n-dimensional space that locally resembles a d-dimensional hyperplane.

Manifold Learning - modeling the maniford on where training instances lie. Based on the manifold assumption - most real-world datasets lie close a lower-dimensional manifold. The constraints of a dataset (what distinguishes, i.e. basketball images from other images of oceans, buildings, etc) enable them to be squeezed.

However, sometimes this backfires - as the dataset is actually in a higher-dimensional manifold than a lower one. Dimensionality reduction speeds up training, but not always leads to a better/simpler solution.

## PCA (Principal Components Analysis)

Generally chooses the axies which minimizes the mean squared distance between the dataset and projection.

### Principal Components

The unit vector that defines the ith axis.

Principle components are found using a standard matrix factorization technique called Singular Value Decomposition (SVD).

Once principle components are identified, the dataset can be projected onto a n-dimensional hyperplane of the first n principle components. It preserves as much variance as possible.

To project the training set onto hyperplane - matrix multiplication of training set matrix X by matrix Wd.

```python
# Copied from book - visualiing PCA

angle = np.pi / 5
stretch = 5
m = 200

np.random.seed(3)
X = np.random.randn(m, 2) / 10
X = X.dot(np.array([[stretch, 0],[0, 1]])) # stretch
X = X.dot([[np.cos(angle), np.sin(angle)], [-np.sin(angle), np.cos(angle)]]) # rotate

u1 = np.array([np.cos(angle), np.sin(angle)])
u2 = np.array([np.cos(angle - 2 * np.pi/6), np.sin(angle - 2 * np.pi/6)])
u3 = np.array([np.cos(angle - np.pi/2), np.sin(angle - np.pi/2)])

X_proj1 = X.dot(u1.reshape(-1, 1))
X_proj2 = X.dot(u2.reshape(-1, 1))
X_proj3 = X.dot(u3.reshape(-1, 1))

plt.figure(figsize=(8,4))
plt.subplot2grid((3,2), (0, 0), rowspan=3)
plt.plot([-1.4, 1.4], [-1.4*u1[1]/u1[0], 1.4*u1[1]/u1[0]], "k-", linewidth=1)
plt.plot([-1.4, 1.4], [-1.4*u2[1]/u2[0], 1.4*u2[1]/u2[0]], "k--", linewidth=1)
plt.plot([-1.4, 1.4], [-1.4*u3[1]/u3[0], 1.4*u3[1]/u3[0]], "k:", linewidth=2)
plt.plot(X[:, 0], X[:, 1], "bo", alpha=0.5)
plt.axis([-1.4, 1.4, -1.4, 1.4])
plt.arrow(0, 0, u1[0], u1[1], head_width=0.1, linewidth=5, length_includes_head=True, head_length=0.1, fc='k', ec='k')
plt.arrow(0, 0, u3[0], u3[1], head_width=0.1, linewidth=5, length_includes_head=True, head_length=0.1, fc='k', ec='k')
plt.text(u1[0] + 0.1, u1[1] - 0.05, r"$\mathbf{c_1}$", fontsize=22)
plt.text(u3[0] + 0.1, u3[1], r"$\mathbf{c_2}$", fontsize=22)
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$x_2$", fontsize=18, rotation=0)
plt.grid(True)

plt.subplot2grid((3,2), (0, 1))
plt.plot([-2, 2], [0, 0], "k-", linewidth=1)
plt.plot(X_proj1[:, 0], np.zeros(m), "bo", alpha=0.3)
plt.gca().get_yaxis().set_ticks([])
plt.gca().get_xaxis().set_ticklabels([])
plt.axis([-2, 2, -1, 1])
plt.grid(True)

plt.subplot2grid((3,2), (1, 1))
plt.plot([-2, 2], [0, 0], "k--", linewidth=1)
plt.plot(X_proj2[:, 0], np.zeros(m), "bo", alpha=0.3)
plt.gca().get_yaxis().set_ticks([])
plt.gca().get_xaxis().set_ticklabels([])
plt.axis([-2, 2, -1, 1])
plt.grid(True)

plt.subplot2grid((3,2), (2, 1))
plt.plot([-2, 2], [0, 0], "k:", linewidth=2)
plt.plot(X_proj3[:, 0], np.zeros(m), "bo", alpha=0.3)
plt.gca().get_yaxis().set_ticks([])
plt.axis([-2, 2, -1, 1])
plt.xlabel("$z_1$", fontsize=18)
plt.grid(True)

plt.show()
```

Previous implementations of PCA require the whole training set to be fit in memory. IPCA, however, can split the training set into mini-batches, for large training sets/online learning.

Alternative is NumPy's memmap - manipulate a large array stored in binary file as if it's entirely in memory. Loads only the data it needs when it needs it.

## Kernel PCA

(See Chapter 5)
Kernel Trick - mathematical technique implicitly mapping instances to a high-dimensional space (feature space), enabling nonlinear classification and regression with SVM. (adding high polynomial degrees without actually adding them to make the model too slow)

This works because a complex nonlinear decision boundary in the original space, becomes a linear one in high-dimensional spaces.

The same works for PCA. Good at preserving clusters of instances.


```python
# Swiss-roll Toy Dataset

from sklearn.datasets import make_swiss_roll
X, t = make_swiss_roll(n_samplbes=1000, noise=0.2, random_state=24)
```

```python
from sklearn.decomposition import KernelPCA

rbf_pca = KernelPCA(n_components = 2, kernel="rbf", gamma=0.04)
X_reduced = rbf_pca.fit_transform(X)
```

```python
# Reducing Swiss Roll to 2d - Linear Kernel, RBF Kernel, Sigmoid Kernel

# From the book's code

from sklearn.decomposition import KernelPCA

lin_pca = KernelPCA(n_components = 2, kernel="linear", fit_inverse_transform=True)
rbf_pca = KernelPCA(n_components = 2, kernel="rbf", gamma=0.0433, fit_inverse_transform=True)
sig_pca = KernelPCA(n_components = 2, kernel="sigmoid", gamma=0.001, coef0=1, fit_inverse_transform=True)

y = t > 6.9

plt.figure(figsize=(16, 4))
for subplot, pca, title in ((131, lin_pca, "Linear kernel"), (132, rbf_pca, "RBF kernel, $\gamma=0.04$"), (133, sig_pca, "Sigmoid kernel, $\gamma=10^{-3}, r=1$")):
    X_reduced = pca.fit_transform(X)
    if subplot == 132:
        X_reduced_rbf = X_reduced
    
    plt.subplot(subplot)
    #plt.plot(X_reduced[y, 0], X_reduced[y, 1], "gs")
    #plt.plot(X_reduced[~y, 0], X_reduced[~y, 1], "y^")
    plt.title(title, fontsize=14)
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot)
    plt.xlabel("$z_1$", fontsize=18)
    if subplot == 131:
        plt.ylabel("$z_2$", fontsize=18, rotation=0)
    plt.grid(True)

plt.show()
```

### Selecting a Kernel and Tuning Hyperparameters

kPCA is unsupervised learning - no obvious way to select best kernel/hyperparameter values. Grid search can be implemented, however. 

Following example:
- Creates 2-step pipeline - using kPCA to reduce to 2-dimensional, then applying logistic regression for classification
- Then uses Grid Search to find the best kernel and gamma values for best accuracy

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('kpca', KernelPCA(n_components=2)),
    ('log_reg', LogisticRegression())
])

param_grid = [{
    'kpca__gamma': np.linspace(0.03, 0.05, 10),
    'kpca__kernel': ['rbf', 'sigmoid']
}]

grid_search = GridSearchCV(clf, param_grid, cv=3)
grid_search.fit(X, y)
```

```python
grid_search.best_params_
```

Another approach is to find the kernel/hyperparameters that yield the lowest reconstruction error. Complex mathematical infinite-dimensional pre-image mean squared distance blah blah blah.

It's achieved by training a supervised regression model, with the projected instances as training set and original instances as targets.

```python
rbf_pca = KernelPCA(n_components=2, kernel='rbf', gamma=0.0433,
                    fit_inverse_transform=True)

X_reduced = rbf_pca.fit_transform(X)
X_preimage = rbf_pca.inverse_transform(X_reduced)

from sklearn.metrics import mean_squared_error
mean_squared_error(X, X_preimage)

# And then use grid search/cross-validation to find the kernel/hyperparameters that minimise the 
# reconstruction error.
```

## LLE

Local Linear Embedding - another nonlinear dimensionality reduction technique.

Manifold learning technique without relying on projections. It works by first measuring how training instances linearly relates to their closest neighbors, and then look for a low-dimensional representation of the training set where the local relationships are best preserved. Very good at unrolling twisted manifolds, without much noise.

```python
from sklearn.manifold import LocallyLinearEmbedding

lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_reduced = lle.fit_transform(X)
```

## Other Dimensionality Reduction Techniques:

- Multidimensional Scaling/MDS - preserving distances
- Isomap - creaitng a graph preserving geodesic distances
- t-Distributed Stochastic Neighbor Embedding (t-SNE) - keeping similar instances close and dissimilar ones apart
- Linear Discriminant Analysis (LDA) - classification, but learns the axis in training
